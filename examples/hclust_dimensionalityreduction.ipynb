# Clustering Jerarquico

```python
import pandas as pd
from scipy.cluster import hierarchy
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
df_nasa = pd.read_csv('https://raw.githubusercontent.com/pokengineer/DataScience/main/datasets/asteroids_nasa.csv')
df_nasa = pd.read_csv('F:/Materias/00-ingreso/taller/regresion/autos.csv')
```

```python
# Seleccionar solo columnas num√©ricas
X = df_nasa.select_dtypes(include='number').drop(['Orbit ID','Neo Reference ID','Name'],axis=1)

# Normalizar las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Transponemos: ahora cada columna original es una "observaci√≥n"
X_t = X_scaled.T

# Clustering jer√°rquico sobre las variables
linkage = hierarchy.linkage(X_t, method='average', metric='euclidean')

# Dendrograma de las columnas
plt.figure(figsize=(14, 6))
dendro = hierarchy.dendrogram(linkage, labels=X.columns, leaf_rotation=90)
plt.title("Dendrograma de Variables - Clustering Jer√°rquico NASA")
plt.xlabel("Variables originales")
plt.ylabel("Distancia")
plt.tight_layout()
plt.show()
```

```python
print("cantidad de columnas: " + str(len(X.columns)))
print( "cantidad de clusters: " + str(len(set(hierarchy.fcluster(linkage, 10, criterion='distance')))) )
```
